```python
import seaborn as sns
import matplotlib.pyplot as plt





# Create a line plot of cholesterol level by age
sns.lineplot(x="age", y="chol", data=ages, hue=None, style=None, markers=False, dashes=True,
             palette="husl", linewidth=2.5, alpha=0.8, drawstyle="steps-post",
             ci="sd", estimator="mean", err_style="bars", err_kws={"capsize": 4})

# Set the plot title and axis labels
plt.title("Cholesterol Level by Age")
plt.xlabel("Age")
plt.ylabel("Cholesterol Level")

# Set the legend title and position
plt.legend(title=None, loc="best")

# Adjust the plot layout and style
sns.set_style("whitegrid")
sns.set(rc={"figure.figsize": (8, 6)})
sns.despine()

# Display the plot
plt.show()
```
